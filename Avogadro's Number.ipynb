```julia
# Brightness threshold for particles in brownian motion:
τ = 180 / 255

# A function to calculate the brightness
luma(c) = .299*c.r + .587*c.g + .114*c.b

# Classify each pixel as black or white.
classify_foreground(img, threshold=τ) =
    map(luma, img.data) .> (threshold)

# Utility function to draw a boolean array
# as a black and white image

function boolimg(array)
    x, y = size(array)
    pixels = fill(RGB(0,0,0), x, y)
    for i=1:x, j=1:y
        c = array[i, j]
        pixels[i, j] = RGB(c, c, c)
    end
    Image(pixels, ["spatialorder" => ["x", "y"]])
end
```

# Finding Blobs

We now make a library for `Point`s and `Blob`s.

```julia
# The Point type
immutable Point
    x::Float64
    y::Float64
end

# Arithmatic of points
+(p::Point, q::Point) = Point(p.x + q.x, q.y + q.y)
/(p::Point, a::Real) = Point(p.x / a, p.y / a)
-(p::Point, q::Point) = Point(q.x - p.x, q.y - p.y)

abssquare(p::Point) = p.x^2 + p.y^2
@vectorize_1arg Point abssquare # make abs work for a vector of points

midpoint(ps::Array{Point}) = reduce(+, Point(0, 0), ps) / length(ps)

# A Blob is a set of points
immutable Blob
    points::Array{Point}
end

# mass of a blob is just the number of points in it
mass(b::Blob) = length(b.points)
# center of the blob is its centroid
center(b::Blob) = midpoint(b.points)
```

Using the above library functions we will now define the blob finding function `findblobs`. `findbolobs` takes a boolean image (an image passed to classify_foreground), and returns a vector of `Blobs` in the image.

```julia

# Helper function to recursively explore a single white region.
# It modifies the visited matrix as it visits a pixel.

function blobpixels!(img, i, j, visited, points=Point[])
    w, h = size(img)
    if i > w || j > h || i < 1 || j < 1
        return points
    end

    if !visited[i,j] && img[i, j]
        visited[i, j] = true
        push!(points, Point(i, j))
        blobpixels!(img, i-1, j, visited, points)
        blobpixels!(img, i+1, j, visited, points)
        blobpixels!(img, i, j-1, visited, points)
        blobpixels!(img, i, j+1, visited, points)
    end
    return points
end

# Find blobs in a boolean matrix.
function findblobs(bitimg)
    blobs = Blob[]
    visited = zeros(Bool, size(bitimg)) # an array of the same size as the pic

    w, h = size(bitimg)
    for i=1:w, j=1:h
        if !visited[i, j] && bitimg[i, j]
            # Not visited, but white. A new blob!
            points = blobpixels!(bitimg, i, j, visited)
            push!(blobs, Blob(points))
        end
    end
    blobs
end

# Filter out blobs that aren't big enough.
keepbeads(blobs, minmass=25) =
    filter(b->mass(b) >= minmass, blobs)
```

```julia
# Twenty different colors for the blobs
colors = distinguishable_colors(20, cchoices=linspace(80, 90, 20), lchoices=linspace(40, 90, 20))

# A utility function to mark blobs on an image.
function markblobs(img, blobs, colormap=Dict)
    img′ = copy(img)
    for (id, blob) in enumerate(blobs)
        for point in blob.points
            img′[point.x, point.y] = colors[id % length(colors)]
        end
    end
    img′
end
```

```julia
find_nearest(p, points) =
    findmin([q - p |> abssquare for q in points])

function trackblobs(prev, next)
    
    mapping = Dict{Int, (Int, Float64)}()
    for (from, p) in enumerate(prev)
        dist, to = find_nearest(p, next)
        mapping[from] = (to, dist)
    end
    mapping
end

function find_displacements(prev, next, delta=25.0)
    disp = Float64[]
    for (from, p) in enumerate(prev)
        d, idx = find_nearest(p, next)
        if d < delta^2
            push!(disp, d)
        end
    end
    disp
end
```

## Estimating self-diffusion constant

The Einstein-Smoluchowski equation states that the random displacement of a bead in one dimension has a Gaussian distribution with mean zero and variance $\sigma^2 = 2D\Delta t$.

Also, given the data we have a way estimate to the variance:

$\sigma^2 = {r_1^2 + r_2^2 + ... + r_n^2 \over 2n}$
 
 

```julia
variance(displacements) =
    sum(displacements) / (2length(displacements))
```

## Estimating Avogadro's Number

The Stokes-Einstein relation asserts that the self-diffusion constant of a spherical particle immersed in a fluid is given by

$D = {kT \over 6\pi\eta\rho}$

where, for our data,

- T = absolute temperature = 297 K (room temperature)
- $\eta$ = viscosity of water = $9.135 \times 10^{-4} Ns/m^2$ (at room temperature)
- $\rho$ = radius of bead = $0.5\times10^{-6}m$

and k is the Boltzmann constant, given by

$k = R / N_A$

Where,
- R is the gas constant: approx. $8.31457 J K^{-1}mol^{-1}$
- $N_A$ is the Avogadro's Number

$\implies N_A = R / k$

```julia
m_per_pixel = 0.175 * 10.0^-6

function all_displacements(run)
    getbeads(i) = framefile(run, i) |> imread |> classify_foreground |> findblobs |> keepbeads
    ds = Float64[]
    prev = [midpoint(b.points) for b in getbeads(0)]
    for i=1:199
        next = [midpoint(b.points) for b in getbeads(i)]
        ds = vcat(ds, find_displacements(prev, next, 25))
        prev = next
    end
    ds
end
```

```julia
@time disps_run1 = all_displacements(1);
```

```julia
function boltzmann(D)
    T = 297
    η = 9.135*10.0^-4
    ρ = 0.5*10.0^-6
    
    6π*η*ρ*D / T
end

function est_avogadro(displacements, pixel=m_per_pixel, Δt=0.5)
    σ² = variance(displacements) * (pixel^2)
    D = σ²/2Δt
    k = boltzmann(D)
    
    R = 8.31457
    k, R/k
end
```

```julia
est_avogadro(disps_run1)
```
