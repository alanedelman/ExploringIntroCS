### BRIGHTNESS

```python
# brightness of a color is the sum of the r,g,b values (stored as float32's)
brightness(c)        =  c.r + c.g + c.b

# brightness of an image bordered with zeros
function brightness(I::Image)
  h,w = size(I)
  b = [brightness(I[i,j]) for i=1:h,j=1:w]
  # borders
  zr = fill(0f0,1,w+2) # zero row
  zc = fill(0f0,h,1)   # zero col
  [zr;   zc b zc ;   zr]
 end
```

### ENERGY

```python
# the 3x3 stencil for energy
function stencil(b)
      xenergy = b[1,1]+2b[2,1]+b[3,1]-b[1,3]-2b[2,3]-b[3,3]
      yenergy = b[1,1]+2b[1,2]+b[1,3]-b[3,1]-2b[3,2]-b[3,3]
      √(xenergy^2 + yenergy^2)
end
    
# energy of an array of brightness values with zero borders all around in 
# and left and right inf borders out
function energy(b)
  h,w = size(b)
  e = [float32(stencil(b[i-1:i+1,j-1:j+1])) for i=2:h-1,j=2:w-1]
  infcol = fill(Inf32,h-2,1)
  [infcol e infcol]
end
```

### LEAST ENERGY FROM BOTTOM TO TOP

```python
#   _   e   _
#  ↙   ↓   ↘   <--directions naturally live between the rows
#  ed  ed  ed   <--find the smallest and add up to e


# Take an array of energies and work up from bottom to top
function least_energy(e)
   # initialize
   h, w = size(e)
   dirs = fill(0, h-1, w-2)
   energy_downward = fill(0f0, h, w-2)
   energy_downward[h,:] = e[h,2:w-1]
   # compute
    ### edown should be the sum
   for i=h-1:-1:1, j=2:w-1
        smallest, dirs[i,j-1] = findmin(energy_downward[i+1,j+[-1, 0, 1]])
        energy_downward[i,j-1] = smallest + e[i,j]  # smallest and add up to e
        end
   energy_downward, dirs
end
```

### GET  SEAM

```python
function alan_get_seam(dirs,j)
  seam = fill(0,1+size(dirs,1))
  seam[1]=j
  for i=1:size(dirs,1)
    seam[i+1] = seam[i] + dirs[i,seam[i]] - 2
  end
  seam
end
```

```python
########### Shashi's original
# brightness of the pixel at (i,j)
# 0.0 outside the image boundary
function getbrightness(img, i, j)
    w, h = size(img)
    # Boundaries
    if i < 1 || i > w || j < 1 || j > h
        return 0.0f0
    end
    # Interior
    return brightness(img[i, j])
end


# energy of the pixel centered at (i,j)
function shashienergy(img, i, j)
    # A B C
    # D E F
    # G H I
      w, h = size(img)
      a, d, g, b, e, h, c, f, i = [getbrightness(img, x, y) for x=i-1:i+1, y=j-1:j+1]
     xenergy = a + 2d + g - c - 2f - i
     yenergy = a + 2b + c - g - 2h - i
    √((xenergy^2) + (yenergy^2))
end
```

```python
# A NextRow is a data structure for each pixel containing the seam direction and the energy
immutable NextRow
    direction::Int8  # Southwest = -1, South = 0, Southeast = 1   (initialize:2 error:3)
    energy::Float32  # cumulative energy from the top 
end 
```

`markminpaths` marks the seams from a given x, y coordinate

```python
howmany = zeros(100,75)

# presume the cumulative energies from above are known

function mapseams!(img, mem, x, y=1)
    # compute NextRow at (x, y)
    w, h = size(img)
    if x > w || x < 1
        return NextRow(3, Inf)
    end
    if y > h || y < 1
        return NextRow(3, 0.0f0)
    end
    if mem[x, y].direction != 2
        return mem[x, y]
    end
    
    
    # compute the energy at row y+1 
    # from columns x-1, x, and x+1
    # and see which is smallest
    # and remember whether it was southwest, south, or southeast

    next = mapseams!(img, mem, x-1, y+1)   # look southwest
    E = next.energy
    next_dir = -1
    for i in [x, x+1]  # look south and southeast
        tmp = mapseams!(img, mem, i, y+1)
        if tmp.energy < E
            E = tmp.energy
            next_dir = i-x
        end
    end
    
    res = NextRow(next_dir, E + energy(img, x, y))
    mem[x, y] = res
    return res
end

function mapseams(img)
    w, h = size(img)
    # Preallocate with a matrix the same size as img, and energy initialized to 0
    m = fill( NextRow(2, 0.0f0), (w, h))
    # compute energies one row at a time
    for i=1:w
        mapseams!(img, m, i, 1) # mark all min-seams
    end
    m
end
```

given a matrix of Next values, figure out a seam starting at (x, 1)

```python
# Return a vector of j values from top to bottom 
# first value is x, and we work from top to bottom
function get_seam(seammap, x)
    w, h = size(seammap)

    path = Array(Int, h)
    for y = 1:h
        path[y] = x
        x += seammap[x, y].direction
    end
    path
end
```

```python
using Color

# For inspection, mark a seam

function markseam(img, seam, color=RGB(1,0,0))
    img′ = copy(img)
    w, h = size(img)
    for j=1:length(seam)
        img′[seam[j], j]=color
    end
    img′
end
```

```python
# get the min-seam
function minseam(seammap)
    _, idx = findmin([pos.energy for pos in seammap[:, 1]])
    get_seam(seammap, idx)
end
```

```python
function carve(img, seam) # Carve out a seam
    w, h = size(img)
    img′ = fill(RGB(0.0f0, 0.0f0, 0.0f0), w-1, h)

    for y=1:h
        i=1
        for x=1:w
            if seam[y] == x
            else
                img′[i, y] = img[x, y]
                i += 1
            end
        end
    end
    Image(img′, {"spatialorder" => ["x", "y"]}) # same spatial ordering as the original
end
```
